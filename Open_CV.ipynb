```python
import cv2 as cv
import cv2_ext as cv3
import datetime
import numpy as np

def sharpening(step, lst):
    if step == 6:
        print("\nФункция увеличивает резкость изображения по заданному внутри фиксированному алгоритму.")
        directory = new_path()
        for i in lst:
            img = cv3.imread(i)
            kernel = np.array([[-1, -1, -1],[-1, 9, -1],[-1, -1, -1]])
            sharpped = cv.filter2D(img, -1, kernel)
            writer(directory, sharpped)

def blurring(step, lst):
    if step == 5:
        print("\nФункция размывает изображение по заданному внутри фиксированному алгоритму.")
        directory = new_path()
        for i in lst:
            img = cv3.imread(i)
            blurred = cv.GaussianBlur(img, ksize=(11, 11), sigmaX =0, sigmaY=0)
            writer(directory, blurred)

def resizing(step, lst):
    if step == 3:
        print("\nВ программу вводится число в процентах больше нуля для изменения размера изображения, c выбором значения интерполяции." + 
              "\nПри этом результат выполнения программы при неадекватно введенных значениях непредсказуем.")
        directory = new_path()
        for i in lst:
            img = cv3.imread(i)
            height, width = img.shape[:2]
            percent = float(input("\nВведите процент изменения и нажмите Enter: "))
            number = int(input("\nВведите цифру равную значению интерполяции." +
            "\n1 - INTER_NEAREST: Интерполяция по методу ближайшего соседа." +
            "\n2 - INTER_LINEAR: Билинейная интерполяция." +
            "\n3 - INTER_CUBIC: Бикубическая интерполяция." + 
            "\n4 - INTER_LANCZOS4: Интерполяция Ланцоша по окрестностям 8×8: "))
            numbers = set([1, 2, 3, 4])
            numbers.add(number)
            if len(numbers) == 4 and percent > 0:
                interpolations = {1: cv.INTER_NEAREST, 2: cv.INTER_LINEAR, 3: cv.INTER_CUBIC, 4:cv.INTER_LANCZOS4}
                size_width = int(width * percent / 100)
                size_height = int(height * percent / 100)
                new_size = (size_width, size_height)
                resized = cv.resize(img, new_size, interpolation=interpolations[number])
                writer(directory, resized)
            else:
                print("\nВведено некорректное значение, повторите ввод")
                return resizing(step, lst)
        

def rotation(step, lst):
    if step == 2:
        print("\nВ программу вводятся градус для поворота изображения 2D и число равное масштабированию." + 
              "\nПоворот производится вдоль центра изображения." + 
              "\nПри этом результат выполнения программы при неадекватно введенных значениях непредсказуем.")
        directory = new_path()
        for i in lst:
            img = cv3.imread(i)
            height, width = img.shape[:2]
            center = (width / 2, height / 2)
            degree = int(input("\nВведите градусы в виде целового числа и нажмите Enter: "))
            scale = int(input("\nВведите масштаб в виде целового числа, где 100 процентов это масштаб 1 к 1, и нажмите Enter: "))
            prepObj = cv.getRotationMatrix2D(center, degree, scale/100)
            rotated = cv.warpAffine(img, prepObj, (width, height))
            writer(directory, rotated)
            

def cropping(step, lst):
    if step == 1:
        print("\nВ программе вводятся координаты(x1, y1, x2, y2) для обрезки." + 
              "\nРазмеры для обрезки заданы по формуле:[Ширина, Высота]/[x1:x2, y1:y2]" +
              "\nПри этом размеры обрезки органичены программно, во избежание выхода за размеры изображения и части ошибок." +
              "\nКоординаты x1/у1 должны быть меньше координатов x2/y2 соответственно." + 
               "\nКоординаты x2/y2 по которым введено значение меньше 1, будут заданы координаты равные 1." )
        directory = new_path()
        for i in lst:
            img = cv3.imread(i)
            height, width = img.shape[:2]
            print(f"\nШирина исходного изображения: {width}, высота: {height}")
            startX = max(0, int(input("\nВведите x1 и нажмите Enter: ")))
            startY = max(0, int(input("\nВведите y1 и нажмите Enter: ")))
            endX = max(min(width, int(input("\nВведите x2 и нажмите Enter: "))), 1)
            endY = max(min(height, int(input("\nВведите y2 и нажмите Enter: "))), 1)
            if startX < endX and startY < endY:
                cropped = img[startY:endY, startX: endX]
                writer(directory, cropped)
            else:
                print("\nВведено некорректное значение, повторите ввод.")
                return cropping(step, lst)
                

def flipping(step, lst):
    if step == 4:
        print("\nВ программу вводится цифра для выбора вида отзеркаливания:" + 
              "\n0 - разворот вокруг горизонтальной оси" + 
              "\n1 - разворот вокруг вертикальной оси" + 
              "\n-1 - разворот по обеим осям одновременно")
        number = int(input("\nВведите цифру и нажмите Enter: "))
        numbers = set([0, 1, -1])
        numbers.add(number)
        if len(numbers) == 3:
            directory = new_path()
            for i in lst:
                img = cv3.imread(i)
                flipped = cv.flip(img, number)
                writer(directory, flipped)
        else:
            print("\nВведено некорректное значение, повторите ввод")
            return flipping(step, lst)
            
def writer(directory, condition):
    print(f"\nИспользуем путь {directory} для сохранения картинки?")
    choice = int(input("\nВведите: 0 - да, 1 - нет и нажмите Enter: "))
    if choice == 0:
        name = input("\nВведите имя картинки, без типа файла и нажмите Enter: ")
        type = input("\nВведите тип формат картинки без точки и иных знаков, например(jpg, png и т.д), будет приведено к нижнему регистру и нажмите Enter: ").strip().lower()
        dt = datetime.datetime.now()
        date = dt.strftime("-%d.%m-%H.%M")
        result = directory + name + date + "." + type
        cv3.imwrite(result, condition)
        print(f"\nДанные сохранены: {result}")
    elif choice == 1:
        new_directory = new_path()
        return writer (new_directory, condition)
    else:
        print("\nВведено некорректное значение, повторите ввод.")
        return writer(directory, condition)
        
def incorrect(step):
    numbers = set([1, 2, 3, 4, 5, 6])
    numbers.add(step)
    if len(numbers) > 6:
        print("\nВведено некорректное значение. Работа программы завершена. Запустите программу заново.")


def new_path():
    directory = input(r"Введите путь по примеру: D:\Users\user\some_file\ на конце с наклонной чертой влево для сохранения картинки/картинок после обработки и нажмите Enter: ").strip()
    return directory

def choise():
    step = int(input("\nВыберите тип операции, введите целое число и нажмите Enter:" + 
    "\n1 - обрезка;" + 
    "\n2 - поворот; " + 
    "\n3 - масштабирование;" + 
    "\n4 - отзеркаливание;" +
    "\n5 - размытие;" +
    "\n6 - увеличение резкости: "))
    return step
try:
    lst = []
    print("\nПрограмма реализовывает простые операции с изображениями с помощью модуля OpenCV с последующим сохранением файла в директории." + 
          "\nДля этого необходимо выбрать тип проводимой операции, указать перечень директорий картинки которых подлежат изменению" + 
          " и запрашиваемую информацию для сохранения обработанных картинок. OpenCV поддерживает ограниченный тип1 файлов." + 
          "\nОткрытые источники упоминают: bmp, dib, jpeg, jpg, png, webp, sr, ras, tiff, tif, растровые и векторные геопространственные данные, поддерживаемые GDAL.")
    quantity = int(input("\nФайл будет один или несколько? Нажмите цифру и нажмите Enter: 1 - если  один файл, 2 - если несколько."))
    if quantity == 1:
        path = input(r"Введите путь по примеру: D:\Users\user\some_file\рис.jpg, с одной наклонной чертой и нажмите Enter: ").strip()
        cv3.imread(path)
        lst.append(path)
        step = choise()
        incorrect(step)
    elif quantity == 2:
         numb = abs(int(input("\nУкажите количество файлов целом числом и нажмите Enter." + 
                    "\nПри этом если будет указано отрицательное число, программа заменит его на положительное.")))
         if numb > 0:
             for i in range(numb):
                path = input(r"Введите путь по примеру: D:\Users\user\some_file\рис.jpg, с одной наклонной чертой и нажмите Enter: ").strip()
                cv3.imread(path)
                lst.append(path)
             step = choise()
             incorrect(step)
         else:
             print("\nВведено некорректное значение. Работа программы завершена. Запустите программу заново.")
    else:
        print("\nВведено некорректное значение. Работа программы завершена. Запустите программу заново.")
        
    cropping(step, lst)
    rotation(step, lst)
    resizing(step, lst)
    flipping(step, lst)
    blurring(step, lst)
    sharpening(step, lst)
 
except cv.error:
    print("\nВведено некорректное значение, ошибка задекларирована OpenCV. Работа программы завершена. Запустите программу заново.")
except ValueError:
    print("\nВведено не число, ошибка задекларирована. Работа программы завершена. Запустите программу заново.")
except FileNotFoundError:
    print("\nВведен некорректный путь, ошибка задекларирована. Работа программы завершена. Запустите программу заново.")
except OSError:
    print("\nВведен недопустимый аргумент, ошибка задекларирована. Работа программы завершена. Запустите программу заново.")
except KeyboardInterrupt:
    print("\nПрограмма прервана пользователем, ошибка задекларирована. Работа программы завершена. Запустите программу заново.")
except Exception as e:
    print(e)
    print("\nВведено некорректное значение, ошибка не задекларирована. Работа программы завершена. Запустите программу заново.")


        
        
        
```
