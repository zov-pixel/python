```python
import os
import re
from pathlib import Path
```

```python
class Coductor:
    def __init__(self):
        super().__init__()
        print("Консольное приложение выполнят функции проводника по файловой системе.\n")
        self.choise()
        
    def choise(self):
        self.step = input(r"Выберите функцию, введите команду(например: ls) и нажмите Enter:" + 
        "\nls — отображение списка файлов и папок в текущей директории;" + 
        "\ncd  — изменение текущей директории; " + 
        "\nmkdir — создание новой папки;" + 
        "\nrmdir  — удаление папки;" +
        "\ntouch  — создание пустого файла;" +
        "\nrm  — удаление файла;" +
        "\ninfo — вывод информации о файле: имя, размер и путь к файлу(без наименования файла)" +
        "\nexit — выход из программы.").lower()
        print("\n")
        match = re.search(r"(ls)|(cd)|(mkdir)|(rmdir)|(touch)|(\brm\b)|(info)|(exit)", self.step)
        if match != None:
            comands = {"ls": "self.ls()", "cd": "self.cd()", "mkdir": "self.mkdir()", "rmdir": "self.rmdir()", "touch": "self.touch()",
                      "rm": "self.rm()", "info": "self.info()", "exit": "self.exit()"}
            print(eval(comands[match[0]]))
        else:
            print("Введена некорректная команда, повторите ввод.\n")
            self.choise()
        return ''
            
        
    def ls(self):
        self.directory = os.getcwd()
        for i in os.listdir(self.directory):
            print(i)
        print("\n")
        self.choise()
        return ''
                
    def cd(self):
        directory = input(r"Введите путь новой директории по примеру: D:\Users\user\some_file или exit - для выхода из программы и нажмите Enter: ").strip().lower()
        if directory == "exit":
            self.exit()
        else:
            if os.path.isdir(directory) == True:
                os.chdir(directory)
                print(f"\nДиректория изменена на текущую: {os.getcwd()}\n")
                self.choise()
            else:
                print("\nВведенный путь директории не существует. Повторите ввод.\n")
                self.cd()
        return ''

    
    def mkdir(self):
        foldername = input(r"Введите наименование новой папки для создания ее в текущей директории или exit - для выхода из программы и нажмите Enter: ").strip()
        if foldername == "exit":
            self.exit()
        else:
            os.mkdir(foldername)
            path = os.getcwd()+ "\\" + foldername
            if os.path.isdir(path) == True:
                print(f"\nСоздана новая папка, путь папки: {path}\n")
                self.choise()
            else:
                print("\nПроизошла ошибка, папка не создана, повторите ввод.\n")
                self.mkdir()
        return ''
            
    def rmdir(self):
        foldername = input(r"Введите наименование папки для удаления ее из текущей директории или exit - для выхода из программы и нажмите Enter: ").strip()
        if foldername == "exit":
            self.exit()
        else:
            path = os.getcwd() + "\\" + foldername
            if os.path.isdir(path) == True:
                if len(os.listdir(path)) == 0:
                    os.rmdir(foldername)
                    print(f"\nПапка с наименованием {foldername} удалена из текущей директории.\n")
                else:
                    print(f"\nПапка с наименованием {foldername} содержит в себе файлы, удалите файлы из нее и повторите команду.\n")
                self.choise()
            else:
                print(f"\nПроизошла ошибка ввода, папки {foldername} не существует, повторите ввод.\n")
                self.rmdir()
        return ''

    
    def touch(self):
        filename = input(r"Введите наименование нового файла для создания ее в текущей директории или exit - для выхода из программы и нажмите Enter: ").strip()
        if filename == "exit":
            self.exit()
        else:
            Path(filename).touch()
            path = os.getcwd() + "\\" + filename
            if os.path.isfile(path) == True:
                print(f"\nСоздан новый файл, путь файла: {path}\n")
                self.choise()
            else:
                print("\nПроизошла ошибка, файл не создан, повторите ввод.\n")
                self.touch() 
        return ''
            
    def rm(self):
        filename = input(r"Введите наименование файла для удаления его из текущей директории или exit - для выхода из программы и нажмите Enter: ").strip()
        if filename == "exit":
            self.exit()
        else:
            path = os.getcwd() + "\\" + filename
            if os.path.isfile(path) == True:
                os.remove(path)
                print(f"\nФайл с наименованием {filename} удален из текущей директории.\n")
                self.choise()
            else:
                print(f"\nПроизошла ошибка ввода, файла {filename} не существует, повторите ввод.\n")
                self.rm()
        return ''
        
    def info(self):
        filename = input(r"Введите наименование файла из текущей директории для просмотра информации или exit - для выхода из программы и нажмите Enter: ").strip()
        if filename == "exit":
            self.exit()
        else:
            path = os.getcwd() + "\\" + filename
            if os.path.isfile(path) == True:
                print(f"\nИмя файла: {os.path.basename(path)}\nРазмер файла: {os.path.getsize(path)}\nПуть к файлу: {os.getcwd()}\n")
                self.choise()
            else:
                print(f"\nПроизошла ошибка ввода, файла {filename} не существует, повторите ввод.\n")
                self.info()
        return ''
        
    def exit(self):
        print("\nПрограмма завершена.")
        return ''

try:
    app = Coductor()
except Exception as e:
    print(f"Произошла ошибка: {e}. Программа завершена. Запустите при необходимости повторно.")
except KeyboardInterrupt as k:
    print(f"Произошла ошибка: {k}. Программа завершена. Запустите при необходимости повторно.\n")

    
            
            
```
