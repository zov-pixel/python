```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

```python
fig, ax = plt.subplots(figsize=(6, 3.5))
x = [i for i in range(0, 10)]
y = [i**2 for i in x]
ax.plot(x, y)
plt.title("Пример линейного графика")
plt.xlabel("ось x")
plt.ylabel("ось y")
plt.grid()
```

```python
np.random.seed(19685689)
fig, ax = plt.subplots(figsize=(6, 3.5))
mu = 0
sigma = 1
x = mu + sigma*np.random.randn(100)
num_bins = 20
n = ax.hist(x, num_bins, density=True, facecolor='C9', edgecolor = "white",  alpha=1)
plt.title("Пример гистограммы", color="C30", size=15, style="italic")
ax.set_xlabel("ось x", color="C20")
plt.ylabel("ось y", color="C20")
ax.tick_params(labelcolor="C11")
plt.text(-2.2, .5, r'$\mu=0,\ \sigma=1$', color='C10')
```

```python
fig, ax = plt.subplots(figsize=(6, 3.5))
category = ['A', 'B', 'C', 'D']
percent = [40, 30, 20, 10]
explode = (0.1, 0.05, 0, 0)
ax.pie(percent, labels=category, startangle=160, explode=explode)
plt.title("Пример круговой диаграммы")
ax.legend(percent, title="Проценты", loc="center left", frameon=False, bbox_to_anchor=(0.7, 0.7), bbox_transform=fig.transFigure)
```
