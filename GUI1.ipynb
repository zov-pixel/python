```python
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showerror, showinfo, showwarning
from isdigit import IsDigit
import tkinter.scrolledtext as st
import random
import time
import unittest
```

```python
class Sorting(tk.Tk):
    '''Класс для сортировки значения чисел'''
    def __init__(self):
        """Устанавливает все необходимые атрибуты для объекта, содержит в себе параметры:
        geometry() - метод, предустановленное значение, принимающие строку с размером экрана пользователя;
        title() - метод, предустановленное значение, принимающие строку с наименованием экрана пользователя;
        resizable() - метод, предустановленное значение, принимающие 2 булева или аналога их чисел для отключения изменения размера экрана пользователя;
        c_widgets() - метод, не принимающий данные, запускает перечень вижджетов внутри себя для экрана пользователя;
        res_time - переменная int, необходимая в дальнейшем для отображения времени сортировки, с предустановленным значением 0;
        len_lst - переменная int, необходимая в дальнейшем для работы с полученными данными от пользователя, с предустановленным значением 0;"""
        super().__init__()
        self.geometry('450x250')
        self.title('Cортировка последовательности чисел')
        self.resizable(0,0)
        self.c_widgets()
        self.res_time=0
        self.len_lst=0
        
    def repeat(self, result):
        """Метод, принимающий булево, True - повторить операцию, False - завершить, используется метод destroy() для уничтожения виджетов"""
        if result == True:
            self.destroy()
            self.__init__()
        else:
            self.destroy()
            
            
    def ascending_quicksort(self, nums):
        """Метод, принимающий список чисел для сортировки по возрастанию. 
        Метод фиксирует время выполнения сортировки и фиксирует его в переменной self.res_time
           Использует методы: time.perf_counter_ns() - для фиксации дата начала 
           и завершение операции в наносекундах и random.choice() для выбора случайного числа в списке.
           Возвращает результат сортировки."""
        start = time.perf_counter_ns()
        result = []
        if len(nums) <= 1:
            result =  nums
        else:
            q = random.choice(nums)
            l_nums = [n for n in nums if n < q]
            e_nums = [q] * nums.count(q)
            b_nums = [n for n in nums if n > q]
            result = self.ascending_quicksort(l_nums) + e_nums + self.ascending_quicksort(b_nums)
        finish = time.perf_counter_ns()
        self.res_time = (finish-start)
        return result
        
    def descending_quicksort(self, nums):
        """Метод, принимающий список чисел для сортировки по убыванию. 
        Метод фиксирует время выполнения сортировки и фиксирует его в переменной self.res_time
           Использует методы: time.perf_counter_ns() - для фиксации дата начала 
           и завершение операции в наносекундах и random.choice() для выбора случайного числа в списке.
           Возвращает результат сортировки."""
        start = time.perf_counter_ns()
        result = []
        if len(nums) <= 1:
            result =  nums
        else:
            q = random.choice(nums)
            l_nums = [n for n in nums if n < q]
            e_nums = [q] * nums.count(q)
            b_nums = [n for n in nums if n > q]
            result = self.descending_quicksort(b_nums) + e_nums + self.descending_quicksort(l_nums)
        finish = time.perf_counter_ns()
        self.res_time = (finish-start)
        return result

    def choose(self, string, selection):
        """Метод ветвления по типу сортировки в зависимости от выбранного значения пользователем. 
        Принимает параметры данные чисел веденные пользователем в виде строки для запуска метода list_with_warning() списка с предупреждениями
        и выбранного пользователем типа сортировки, в случае отсутсвия выбора и наличия чисел для сортировки, предупреждает о сортировки по умолчанию
        по возрастанию. При отсутствии данных выдает соответствующие предупреждения. 
        При наличии данных выполняет соответствующую выбранному типу сортировку.
        Возращает результат операции -0, в случае проблем и результат выполнения методов сортировки в случае успеха.
        """
        lst = self.list_with_warning(string)
        res = 0
        if len(lst) == 0:
            if selection == "":
                showwarning(title="Warning", message = "Не выбран тип сортировки.")
            res = 0
        elif selection == "cортировка по возрастанию":
            res = self.ascending_quicksort(lst)
        elif selection == "сортировка по убыванию":
            res = self.descending_quicksort(lst)
        else:
            showwarning(title="Warning", message = "Не выбран тип сортировки, cортировка по возрастанию выполнена по умолчанию")
            res = self.ascending_quicksort(lst)
        return res   
        
    
    def results(self, string, selection, frame_result, login_button, combobox, numbers_input):
        """Метод показывает результат выполнения программы пользователю на экране. 
        Принимает виджеты frame_result, login_button, combobox, numbers_input для перерисовки данных, 
        в том числе блокировки для корректировки пользователем.
        Изменяет размер экрана пользователя методом geometry() принимающий значение в виде строки.
        Отображает время сортировки пользователю и результат, в том числе в случае пустых значений при запуске метода choose(), 
        записанный в переменную res.
        Устанавливает значение сортировки по умолчанию, при пустом значении.
        В методе появляются виджеты Close(закрыть) и Repeat(повторить), позволяющий пользователю 
        выбрать порядок действий дальнейших после отбражения сортировки, задействов метод repeat()"""
        label_result = st.ScrolledText(frame_result, font="helvetica 13", foreground="#004D40", height=2)
        label_result.pack(side=tk.TOP,anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)   
        label_result.pack(side=tk.TOP,anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)     
        res = self.choose(string, selection)
        
        if res == 0 and selection=="":
            res = "Содержимое введеных данных не подошло.\nНе выбран тип сортировки."
            label = ttk.Label(frame_result, text= f'Время сортировки не показывается.', font='helvetica 10', relief = "solid")
        elif res == 0:
            res = "Содержимое введеных данных не подошло.\nСортировка не выполнена."
            label = ttk.Label(frame_result, text= f'Время сортировки не показывается.', font='helvetica 10', relief = "solid")
    
        else:
            if selection == "":
                combobox.current(0)
            label = ttk.Label(frame_result, text= f'Сортировка по счетчику производительности\nвыполнена за: {self.res_time:_} наносекунд', font='helvetica 10', relief = "solid")
  
        combobox.config(state=["disabled"])
        numbers_input.config(state=["disabled"])
        label_result.insert(tk.INSERT, res)
        self.geometry('500x500')
        login_button.config(state=["disabled"])
        label.pack(side=tk.TOP,anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)
        button_close=ttk.Button(frame_result, text="Close", command=lambda: self.repeat(False))
        button_close.pack(side=tk.LEFT,anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)
        button_repeat=ttk.Button(frame_result, text="Repeat", command=lambda: self.repeat(True))
        button_repeat.pack(side=tk.LEFT,anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)

    
    def get_list_of_numbers(self, text):
        """Метод принимающий введенные числа пользователем в виде строки. 
        Разделяет строку по запятой и проверяет является ли элемент int или float, используя пакет ISDigit.
        Если элемент попадает под требование, то добавляет его в возвращаемый список. 
        Фиксирует длину первоначального списка по разбиению в переменную self.len_lst"""
        digits = IsDigit(allow_floats=True, allow_ints=True)
        lst = text.split(",")
        final = []
        for i in lst:
            i = i.strip()
            if digits.is_digit((i)) != False:
                final.append(int(i) if i.isdigit() == True else float(i))
        self.len_lst = len(lst)
        return final
        
    def list_with_warning(self, text):
        """Метод, показывающий предупреждение, если числа введенные пользователем не соответствует необходимым,
        сравнивая длину списка полученного методом get_list_of_numbers() и self.len_lst.
        Возвращает список полученный методом get_list_of_numbers().
        """
        lst = self.get_list_of_numbers(text)
        if len(lst) != self.len_lst:
            showwarning(title="Warning", message="Необходимо было ввести цифры с типом int и float,"+
                        " через запятую, пробелы справа или слево от цифры до запятой допустимы, цифры неподходящие под этот критерий"
                        + " не будут включены в список для сортировки, недопустимо пустое значение.")
        return lst
                    
            
        
    def c_widgets(self):
        """Метод, отрисовывающий первоначальный экран с предложением пользователю ввести числа для сортировки через запятую и 
        выбрать тип сортировки, запустив программу нажав кнопку старт. Метод запускает метод results(), 
        передаваемые виджеты для изменения и данные введенные пользователем."""
        style = ttk.Style()
        style.configure(".", font="helvetica 13", foreground="#004D40", padding=8, background="#B2DFDB")
        frame_first = tk.Frame(self)
        frame_first = tk.LabelFrame(self, text = "Введите числа для сортировки, через запятую:", foreground="#004D40", font="helvetica 13")
        frame_first.pack(side=tk.TOP, anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)
                
        numbers_input = st.ScrolledText(frame_first, font="helvetica 13", foreground="#004D40", height=1)
        numbers_input.pack(side=tk.TOP, anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)

        frame_second = ttk.Frame(self)
        frame_second = tk.LabelFrame(self, text ='Выберите тип сортировки:', foreground="#004D40", font="helvetica 13")
        frame_second.pack(side=tk.TOP,anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)
        
        t =tk.StringVar()
        
        entry = ttk.Entry(self, textvariable=t)
        type_of_sorting = ["cортировка по возрастанию", "сортировка по убыванию"]
        combobox = ttk.Combobox(frame_second, values = type_of_sorting, textvariable=t, state="readonly")
        combobox.pack(side=tk.TOP,anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)
    
        frame_buttom = tk.Frame(self)
        frame_buttom.pack(side=tk.TOP,anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)
        login_button=ttk.Button(frame_buttom, text="Start", command=lambda: self.results(numbers_input.get('1.0', tk.END), 
                                                                                 entry.get(), frame_result, login_button, combobox, numbers_input))
        login_button.pack(side=tk.TOP,anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)
        
        self.frame_result = tk.Frame(self)
        frame_result = tk.LabelFrame(self, text ='Результаты:', foreground="#004D40", font="helvetica 13")
        frame_result.pack(side=tk.TOP,anchor=tk.NW, fill=tk.X, expand=True, padx=5, pady=5)
        
        
        
 
class SimpleisticTest(unittest.TestCase):
    """Класс для проведения тестов на соответствие / не соответствие переданных и полученных данных по методам сортировки."""
    def test_equals_ascending_quicksort(self):
        """Метод сверяющий равность значений при сортировке по возрастанию переданного и полученного списком."""
        self.assertEqual(sorting.ascending_quicksort([1, 2, 3]), [1,2,3])
    def test_no_equals__ascending_quicksort(self):
        """Метод сверяющий неравность значений при сортировке по возрастанию переданного и полученного списком."""
        self.assertNotEqual(sorting.ascending_quicksort([1, 2, 3]), [3,2,1])
    def test_equals_descending_quicksort(self):
        """Метод сверяющий равность значений при сортировке по убыванию переданного и полученного списком."""
        self.assertEqual(sorting.descending_quicksort([1, 2, 3]), [3,2,1])
    def test_no_equals__descending_quicksort(self):
        """Метод сверяющий неравность значений при сортировке по убыванию переданного и полученного списком."""
        self.assertNotEqual(sorting.descending_quicksort([1, 2, 3]), [1,2,3])
    def test_no_equals__get_list_of_numbers(self):
        """Метод, сверяющий получаемый список на равность значений, полученные при обработке чисел, введенных пользователем
        и сформированного списка для сортировки."""
        self.assertEqual(sorting.get_list_of_numbers('1, 2/0, 2.0, 4, 6, 1k, l'), [1, 2.0, 4, 6])
    def test_equals_choose_ascending(self):
        """Метод сверяющий равность значений при выборе типа сортировки по возрастанию, проверяет равность переданного и полученного списком."""
        self.assertEqual(sorting.choose("1, 2, 3, 4", "cортировка по возрастанию"), [1, 2, 3, 4])
    def test_equals_choose_descending(self):
        """Метод сверяющий равность значений при выборе типа сортировки по убыванию, проверяет равность переданного и полученного списком."""
        self.assertEqual(sorting.choose("1, 2, 3, 4, 6", "сортировка по убыванию"), [6, 4, 3, 2, 1])
        
class TestNoRaiseValue(unittest.TestCase):
    """Класс для проведения тестов на исключения при использовании методов сортировки."""
    def test_no_exeption(self):
        """Метод для проведения тестов на исключения при использовании методов сортировки, при ошибках выдается сообщение."""
        try:
            sorting.ascending_quicksort([1, 2, 3])
            sorting.descending_quicksort([1, 2, 3])
            sorting.get_list_of_numbers('1, 2/0, 2.0, 4, 6, 1k, l')
            sorting.choose("1, 2, 3, 4", "cортировка по возрастанию")
            sorting.choose("1, 2, 3, 4, 6", "сортировка по убыванию")
          
        except Exception:
            self.fail("Исключение возникло. Такого не должно было случиться.")

sorting = Sorting()
result = 0
if __name__=="__main__":
    suite1 = unittest.TestLoader().loadTestsFromTestCase(SimpleisticTest)
    runner1 = unittest.TextTestRunner(verbosity=1).run(suite1)
    suite2 = unittest.TestLoader().loadTestsFromTestCase(TestNoRaiseValue)
    runner2 = unittest.TextTestRunner(verbosity=1).run(suite2)
    result1 = runner1.wasSuccessful()
    result2 = runner2.wasSuccessful()
    result = result1+result2
    
if result == 2:
    sorting.mainloop()
else:
    print("Не сегодня. Программу надо доработать.")

```
