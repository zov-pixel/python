```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(6, 3.5))
x = [i for i in range(0, 10)]
y = [i**2 for i in x]
d = {"ось x": x, "ось у": y}
df=pd.DataFrame(data=d)
ax =sns.lineplot(data=df, x="ось x", y="ось у").set(title="Пример линейного графика")
sns.set_style("darkgrid")
```

```python
np.random.seed(19685689)
fig, ax = plt.subplots(figsize=(6, 3.5))
mu = 0
sigma = 1
res = mu + sigma*np.random.randn(100)
d = {"ось x": res}
df=pd.DataFrame(data=d)
num_bins = 20
color = sns.color_palette('hls')
ax = sns.histplot(data=df, x="ось x", bins=num_bins, color=color[3]).set(title="Пример гистограммы")
plt.ylabel("ось y")
plt.legend(df, loc="upper left")
sns.set_style("white")
```

```python
fig, ax = plt.subplots(figsize=(6, 3.5))
category = ['A', 'B', 'C', 'D']
percent = [40, 30, 20, 10]
pallete_color = sns.color_palette('dark')
explode = (0.05, 0, 0, 0)
ax.pie(percent, labels=category, colors=pallete_color, explode=explode, autopct="%.0f%%")
ax.set_title("Пример круговой диаграммы")
```
